```python
from urllib.request import urlopen
import pandas as pd
from bs4 import BeautifulSoup
import webbrowser
import numpy as np
import re
```

#### 종목뽑기

```python
url ='https://lab.donutz.co/krx/products'
resultXML=urlopen(url)
result=resultXML.read()
soup=BeautifulSoup(result,'html.parser')
soup_string = str(soup)
regex = re.compile(r'(\d{6})')
code = regex.findall(soup_string)
code = [i for i in code]
```

#### 입력하기

```python
#회사 종목
code='005930' 

#시작 날짜
start_date='20101101'

#끝 날짜
end_date = '20201116'
```

#### 한개의 종목 크롤링 코드

```python
url ='https://lab.donutz.co/krx/products/'+code+'?sdate='+start_date+'&edate='+end_date
resultXML=urlopen(url)
result=resultXML.read()
soup=BeautifulSoup(result,'html.parser')
soup_string = str(soup)

def clean(data):
    text = re.sub('[a-zA-Z-=+#/\?^$@*\"※~&%ㆍ!_』:\\‘|\(\)\[\]\<\>`\'{}…》]', '', data)
    return text

text=re.split('[,]+',clean(soup_string))
information=[]
for i in text:
    information.append(i)
    
data=pd.DataFrame(np.array(information).reshape(-1,15))
data.columns=['date','open','high','low','close','trading_volume','score','index','probability','l1','l2','l3','l4','lgap','lrate']
```
